Clustering Using K_Mean


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = np.array([
     [15,39],[16,81],[18,77],[19,40],[12,34],[13,78],[34,76],[21,56],[31,89],[15,64],[17,48],[11,89],[25,66],[19,99]
])
```

```python
kmean = KMeans(n_clusters=2 , random_state=55)
kmean.fit(data)
labels = kmean.labels_
```

```python
plt.scatter(data[:,0],data[:,1],c=labels,cmap='viridis')
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],s=30,c='black',marker='X')
plt.xlabel("Annual Income")
plt.ylabel("Spending score %")
plt.title("customer segmentation using kmean")
plt.show()
```
